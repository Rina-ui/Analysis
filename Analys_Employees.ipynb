                                                                    Analyse par agrégation

Nous allons analyser dans un premier temps les profiles des employees qui partent

```python
# Quel département a le taux d'attrition le plus élevé ?
taux = data_employees.groupby('Department')['Attrition'].sum()
print(taux)
```

D'apres l'analyse le departement ayant le taux d;attrition le plus eeleve est le sdepartement de Research & Development

```python
# Les employés qui font des heures sup (`OverTime`) partent-ils plus ?
hours_employee = data_employees.groupby('OverTime')['Attrition'].mean()
print(hours_employee)
```

                                                                Interpretation du resultat

Ici on voit clairement que 30,53 des employees partent en faisant des heures supplementaires et que 10,44 des employees faisant des heures sup ne partent pas

```python
# Quel est l'âge moyen des employés qui quittent vs qui restent ?
mean_age = data_employees.groupby('Attrition')['Age'].mean()
print(mean_age)
```

                                                                    Interpretation du resultat
Nous remarquons que l'age moyen des employees qui partent est de 33-34 et moins et ceux qui restent de 37-38 ans
En conclusion les jeunes partenet le plus

```python
# Le niveau de salaire (`MonthlyIncome`) influence-t-il le départ
salary_income = data_employees.groupby('Attrition')['MonthlyIncome'].mean()
print(salary_income)
```

                                                                    Interpreation du resultat
Nous voyons clairement que les employees qui partent gagnent au plus 4787 de salaire a contrarion de ceux qui restent qui gagnent au moins en 6000 et plus
En conclusion ceux qui partenet gagnent moins que ceux qui restent
